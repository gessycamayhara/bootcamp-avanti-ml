## Importando bibliotecas

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Plotando gráficos simples

```python
fig, ax = plt.subplots() # Figura com um único eixo
ax.plot([1,2,3,4],[1,4,2,3])
ax.set_title("Teste")
ax.set_xlabel("X axis label")
ax.set_ylabel("Y axis label")
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3], [10, 20, 15])  # plota os dados
ax.set_title("Meu Gráfico")
ax.set_xlabel("Eixo X")
ax.set_ylabel("Eixo Y")

plt.show()
```

```python
fig, axs = plt.subplots(2,2) # Uma figura com uma grade 2x2 de eixos.
```

```python
# Figura sem eixos na esquerda e dois na direita
fig, axs = plt.subplot_mosaic([['left', 'right_top'], ['left', 'right_bottom']])
```

```python
# Convertendo um numpy matrix para numpy array
b = np.matrix([[1,2],[3,4]])
b_asarray = np.asarray(b)
print(b_asarray)
```

```python
np.random.seed(19680801)
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

```python
# Dados
dados = [24, 24, 25, 25, 27, 30, 32, 32, 35, 43]

# Criar o boxplot
plt.figure(figsize=(3, 4))
box = plt.boxplot(dados, vert=True, patch_artist=True, showfliers=True)

# Personalizar cor da caixa
for patch in box['boxes']:
    patch.set_facecolor('white')

# Títulos e rótulos
plt.ylabel('Valores')
plt.title('Boxplot com Quartis e Outlier')
plt.grid(True)

# Mostrar gráfico
plt.show()
```
